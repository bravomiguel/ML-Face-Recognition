### Parameter Options

* I = path to the image you want to test the function on
* featureType = type of feature extraction to be performed on detected faces. Supported feature types: 'pca', 'surf'
* classifierType = type of classifier to use when identifying faces. Supported classifiers: 'cnn', 'svm', 'mlp'
* creativeMode = renders detected faces in an artistic style, when switched on. Supported values: 0 (off), 1 (on)
